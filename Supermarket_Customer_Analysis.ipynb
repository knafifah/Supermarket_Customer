## **1.1 Background**

Edeka is one of the largest supermarket chains in Europe, known for offering a wide range of high-quality food, beverages, and household products. The supermarket is known for its focus on fresh produce and local specialties. Customers can shop for everything from fresh produce and meat to gold products. ## **1.2 Business Problem**

The supermarket has been running multiple marketing campaigns aimed at boosting sales and increasing customer retention. There are six campaigns designed to appeal to different customer demographics, including age groups, marital status, and household composition. However, the supermarket has noticed that some campaigns are more successful than others and the acceptance rate, the percentage of customers accepted the offer in each campaign, varies widely.

The supermarket aims to understand which customer segments are most responsive to the marketing campaigns and what product generates the most sales. By doing so, the supermarket can tailor future marketing campaigns efforts, improve customer engagement, and optimize profit.

## **1.3 Business Objective**

The supermarket primary goal is to boost sales and its campaigns by identifying the key that are most likely to respond positively to each campaign. In order to do that, the supermarket needs to evaluate each campaign's effectiveness, identify customer segments, analyze spending behaviour.

- Identifying which product brings in the most revenue
- Evaluating the acceptance rate for each campaign
- Identifying which customer groups are most responsive to the campaigns
- Identifying the key factors influencing customers spending behaviour or campaigns engagement
- Understanding if segmentation influences spending behaviour or campaign acceptance

> Normality Test of Income

Hypotheses:
- H0: Income is normally distributed
- H1: Income is not normally distributed

```python
from scipy.stats import shapiro
stats, pvalue = shapiro(df[df['Income'].notna()]['Income'])

if pvalue < 0.05:
    print(f'pvalue = {pvalue} < 0.05\nReject H0 --> Income is not normally distributed.')
else:
    print(f'pvalue = {pvalue} > 0.05\nFail to reject H0 --> Income is normally distributed.')
```

From normality test, we got that the `Income` variable is not normally distributed and we can see from the plot above that the frequency distribution of Income is right skewed, so we will fill the missing values with median.

```python
from scipy.stats import kruskal
```

We will use hypothesis testing to evaluate the mean rank of Income across multiple categories, such as Education, Age Category, and Marital Status.

> Hypotheses Testing for Mean Rank of Income (Education)

Hypotheses:
- H0: The mean rank of income customers with basic education = The mean rank of income customers with graduation education = The mean rank of income customers with Master degree = The mean rank of income customers with PhD degree
    
    --> The mean rank of income for each group is the same

- H1: The mean rank of income customers with basic education ≠ The mean rank of income customers with graduation education = The mean rank of income customers with master education = The mean rank of income customers with phd education
    
    --> There's at least one group whose mean rank of income is different

```python
stats, pvalue = kruskal(df[df['Income'].notna()].loc[df['Education'] == 'Basic', 'Income'],
                         df[df['Income'].notna()].loc[df['Education'] == 'Graduation', 'Income'],
                         df[df['Income'].notna()].loc[df['Education'] == 'Master', 'Income'],
                         df[df['Income'].notna()].loc[df['Education'] == 'PhD', 'Income'])
pvalue

if pvalue < 0.05:
    print(f'pvalue = {pvalue} < 0.05\nReject H0 --> There\'s at least one group whose mean rank of income is different.')
else:
    print(f'pvalue = {pvalue} > 0.05\nFail to reject H0 --> The mean rank of income for each group is the same.')
```

> Hypotheses Testing for Mean Rank of Income (Age Category)

Hypotheses:
- H0: The mean rank of income young adult customers = The mean rank of income adult customers = The mean rank of income middle-aged customers = The mean rank of income elderly customers
    
    --> The mean rank of income for each group is the same

- H1: The mean rank of income young adult customers ≠ The mean rank of income adult adult customers = The mean rank of income middle-aged customers = The mean rank of income elderly customers
    
    --> There's at least one group whose mean rank of income is different

```python
stats, pvalue = kruskal(df[df['Income'].notna()].loc[df['Age_Category'] == 'Young Adult', 'Income'],
                         df[df['Income'].notna()].loc[df['Age_Category'] == 'Adult', 'Income'],
                         df[df['Income'].notna()].loc[df['Age_Category'] == 'Middle-Aged', 'Income'],
                         df[df['Income'].notna()].loc[df['Age_Category'] == 'Elderly', 'Income'])
pvalue

if pvalue < 0.05:
    print(f'pvalue = {pvalue} < 0.05.\nReject H0 --> There\'s at least one group whose mean rank of income is different.')
else:
    print(f'pvalue = {pvalue} > 0.05.\nFail to reject H0 --> The mean rank of income for each group is the same.')
```

> Hypotheses Testing for Mean Rank of Income (Marital Status)

Hypotheses:
- H0: The mean rank of income single customers = The mean rank of income married customers = The mean rank of income divorced customers = The mean rank of income widowed customers
    
    --> The mean rank of income for each group is the same

- H1: The rank of income single customers ≠ The rank of income married customers = The rank of income divorced customers = The rank of income widowed customers
    
    --> There's at least one group whose mean rank of income is different

```python
stats, pvalue = kruskal(df[df['Income'].notna()].loc[df['Marital_Status'] == 'Single', 'Income'],
                         df[df['Income'].notna()].loc[df['Marital_Status'] == 'Married', 'Income'],
                         df[df['Income'].notna()].loc[df['Marital_Status'] == 'Divorced', 'Income'],
                         df[df['Income'].notna()].loc[df['Marital_Status'] == 'Widow', 'Income'])
pvalue

if pvalue < 0.05:
    print(f'pvalue = {pvalue} < 0.05\nReject H0 --> There\'s at least one group whose mean rank of income is different.')
else:
    print(f'pvalue = {pvalue} > 0.05\nFail to reject H0 --> The mean rank of income for each group is the same.')
```

We have statistically tested the mean rank of income by some categorical variables, such as age category, education, and marital status and found out that the mean rank of income by age category and education is different. As a result, we will fill the missing values in `Income` column by its median, grouped by age category and education.

```python
df.groupby(['Age_Category', 'Education'])['Income'].median()
```

```python
df['Income'] = df.groupby(['Age_Category', 'Education'], sort=False)['Income'].transform(lambda x: x.fillna(x.median()))
```

After examining the overall acceptance and rejection rates, let's dive into the details of each campaigns to gain a deeper understanding of which ones are underperforming and which are outperforming. First, we will perform a statistical analysis using the chi-square test to determine if there is a significant difference in campaign performance.

> Hypothesis Testing

Hypotheses:
- H0: There's no significant difference in the performance of the campaigns
- H1: There's a significant difference in the performance of the campaigns

```python
from scipy.stats import chi2_contingency

CmpCount = pd.DataFrame({
    '1st Campaign': [df['AcceptedCmp1'].value_counts()[0], df['AcceptedCmp1'].value_counts()[1]],
    '2nd Campaign': [df['AcceptedCmp2'].value_counts()[0], df['AcceptedCmp2'].value_counts()[1]],
    '3nd Campaign': [df['AcceptedCmp3'].value_counts()[0], df['AcceptedCmp3'].value_counts()[1]],
    '4nd Campaign': [df['AcceptedCmp4'].value_counts()[0], df['AcceptedCmp4'].value_counts()[1]],
    '5nd Campaign': [df['AcceptedCmp5'].value_counts()[0], df['AcceptedCmp5'].value_counts()[1]],
    'Last Campaign': [df['Response'].value_counts()[0], df['Response'].value_counts()[1]]
    })

chi2, p, dof, expected = chi2_contingency(CmpCount)

if p < 0.05:
    print(f'''
    pvalue = {pvalue} < 0.05
    Reject H0 --> There\'s a significant difference in the performance of the campaigns.
    ''')
else:
    print(f'''
    pvalue = {pvalue} > 0.05
    Fail to reject H0 --> There\'s no significant difference in the performance of the campaigns.
    ''')
```

```python
CmpCount = CmpCount.reset_index().rename(columns={'index':'AcceptedOrNot'})
CmpCount['AcceptedOrNot'] = CmpCount['AcceptedOrNot'].map({0:'Rejected', 1:'Accepted'})
CmpCount
```

```python
CmpCount = CmpCount.melt(id_vars='AcceptedOrNot', var_name='Campaign', value_name='Count')
CmpCount
```
